```python
url = "https://www.motorcyclespecs.co.za/model/triu/triumph-street-triple-rs-22.html"
url2 = "https://www.motorcyclespecs.co.za/model/ducati/ducati_scrambler_desert_sled_21.html"
url3 = "https://www.motorcyclespecs.co.za/model/bmw/bmw-r100-7-78.html"
```

```python
# https://oxylabs.io/blog/python-web-scraping
import requests
response = requests.get(url3)
print(response.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'lxml')
# https://medium.com/geekculture/web-scraping-tables-in-python-using-beautiful-soup-8bbc31c5803e
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/#a-regular-expression
soup
```

```python
import re
hook = soup.find_all(string=re.compile(".*Make Model.*"))
assert(len(hook)==1)  # Assert we find one and exactly one NavigatableString containing "Make Model"
hook = hook[0]
table = hook.find_parent('table')
table
```

```python
rows = table.find_all('tr')
len(rows)
```

```python
rows[0].find_all('td')[1]
```

```python
rows[0].find_all('td')[0].text.strip()
```

```python
for row in rows:
    cols = row.find_all('td')
    assert(len(cols) == 2) # We expect two columns in each row, which we treat as a name:value pair
    name = cols[0].get_text().strip()
    value = cols[1].get_text().strip().replace("\t", "")
    print(name, ":", value, "\n")
```
